### Récupérons le contenu total dans les 3 langues

```python
def get_all_text_to_txt(lang, soup):
    list_t = []
    for tag in soup.find_all('rubrique', {"lan": lang}):
        t = tag.find('texte').get_text()
        list_t.append(t) 
    return list_t
```

```python
co_txt = []
fr_txt = []
it_txt = []

for file in xml_list:
    
    with open(file, 'r') as f:
        content = f.read()
    soup = bs.BeautifulSoup(content, 'xml')
    
    co_l = get_all_text_to_txt('co', soup)
    co_txt.extend(co_l)
    
    fr_l = get_all_text_to_txt('fr', soup)
    fr_txt.extend(fr_l)
    
    it_l = get_all_text_to_txt('it', soup)
    it_txt.extend(it_l)
```

### Faisons un nettoyage très basique du contenu

```python
co_txt = ' '.join(co_txt)
fr_txt = ' '.join(fr_txt)
it_txt = ' '.join(it_txt)
```

```python
co_txt = re.sub(r'\n', '', co_txt)
fr_txt = re.sub(r'\n', '', fr_txt)
it_txt = re.sub(r'\n', '', it_txt)
```

```python
co_txt = co_txt.lower()
fr_txt = fr_txt.lower()
it_txt = it_txt.lower()
```

```python
co_txt = unidecode.unidecode(co_txt)
fr_txt = unidecode.unidecode(fr_txt)
it_txt = unidecode.unidecode(it_txt)
```

### Maintenant, faisons les statistiques exploratoires

```python
tokenizer = RegexpTokenizer(r'\w+')
co_token = tokenizer.tokenize(co_txt)
fr_token = tokenizer.tokenize(fr_txt)
it_token = tokenizer.tokenize(it_txt)
```

    2. Calculons les hapax

```python
import pandas as pd
```

```python
co_freq= dict(nltk.FreqDist(co_token))
fr_freq= dict(nltk.FreqDist(fr_token))
it_freq= dict(nltk.FreqDist(it_token))
```

```python
co_freq['a']
```

```python
def hapax_counter(freq_tokens):
    list_hapax = []
    for k, v in freq_tokens.items():
        if 1 == v:
            list_hapax.append(k)
    return list_hapax
```

```python
co_hapax = hapax_counter(co_freq)
fr_hapax = hapax_counter(fr_freq)
it_hapax = hapax_counter(it_freq)
```

```python
distri_hapax_co = len(co_hapax)/len(co_token)
distri_hapax_fr = len(fr_hapax)/len(fr_token)
distri_hapax_it = len(it_hapax)/len(it_token)
```

```python
rapport_hapax = pd.DataFrame(data={"Corpus":["Corsu", "Français", "Italiano"],"Distribution des hapax":[distri_hapax_co, distri_hapax_fr, distri_hapax_it],"Longueur du corpus":[len(co_token), len(fr_token), len(it_token)]})
rapport_hapax = rapport_hapax.set_index('Corpus')
rapport_hapax
```

    3. Loi de Zipf, sur corpus plus réduit

        3.1. Représentation des fréquences de mots

```python
def duplicates_counter(freq_tokens):    
    duplicates = []
    for k, v in freq_tokens.items():
        if 1 < v:
            duplicates.append(k)
    return duplicates
```

```python
co_dup = duplicates_counter(co_freq)
fr_dup = duplicates_counter(fr_freq)
it_dup = duplicates_counter(it_freq)
```

```python
co_voc = co_dup + co_hapax
co_voc = sorted(co_voc)

fr_voc = fr_dup + fr_hapax
fr_voc = sorted(fr_voc)

it_voc = it_dup + it_hapax
it_voc = sorted(it_voc)
```

        3.2. Représentation de la loi de Zipf

```python
def logarithm_ranks(freq_token):
    rank_sort = list(freq_token.keys())
    logs_rank = []
    n = 1
    for i in rank_sort:
        logs_rank.append(math.log(n))
        n = n+1
    return logs_rank

def logarithms_values(freq_token):
    freq_sort = list(freq_token.values())
    logs = []
    for i in freq_sort:
        logs.append(math.log(i))
    return logs
```

```python
co_logra = logarithm_ranks(co_freq)
co_logs = logarithms_values(co_freq)

plt.scatter(co_logra, co_logs,color='red', marker='.')
plt.plot([0,11], [11,0],color = 'black')
plt.title("Corsican corpus according to Zipf Law (log scale)", fontname="Lato")
plt.ylabel("log(freq)", fontname="Lato")
plt.xlabel("log(rank)", fontname="Lato")
plt.show()
```

```python
fr_logra = logarithm_ranks(fr_freq)
fr_logs = logarithms_values(fr_freq)

plt.scatter(fr_logra, fr_logs,color='orange', marker='.')
plt.plot([0,10], [10.5,0],color = 'black')
plt.title("French corpus according to Zipf Law (log scale)",fontname="Lato")
plt.ylabel("log(freq)",fontname="Lato")
plt.xlabel("log(rank)",fontname="Lato")
plt.show()
```

```python
it_logra = logarithm_ranks(it_freq)
it_logs = logarithms_values(it_freq)

plt.scatter(it_logra, it_logs,color='green', marker='.')
plt.plot([0,10.2], [9.9,0],color = 'black')
plt.title("Italian corpus according to Zipf Law (log scale)",fontname="Lato")
plt.ylabel("log(freq)",fontname="Lato")
plt.xlabel("log(rank)",fontname="Lato")
plt.show()
```
