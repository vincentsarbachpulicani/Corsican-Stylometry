```python
from nltk.tokenize import RegexpTokenizer
from nltk import FreqDist
import glob
import matplotlib.pyplot as plt
import re
```

```python
alessa = "./test2/Alessa/"
pe_rocca = "./test2/Rocca/"
appinz = "./test2/Appinza/"

paths = [alessa, pe_rocca, appinz]
names = ["M._Alessandri_di_Chidazzu", "Petru_Rocca", "Martinu_Appinzapalu"]
```

```python
index = 0

for path in paths:
    files = glob.glob(f"{path}/*.txt")
    
    text = []
    for file in files:
        if re.search(r"_co_", file):
            with open(file, 'r') as f:
                content = f.readlines()
            text.extend(content)
    
    tokenizer = RegexpTokenizer('\w+')
    text = ' '.join(text)
    token = tokenizer.tokenize(text)
    
    freq = FreqDist(token)
    
    freq.plot(25,title=names[index])
    
    index += 1
```
