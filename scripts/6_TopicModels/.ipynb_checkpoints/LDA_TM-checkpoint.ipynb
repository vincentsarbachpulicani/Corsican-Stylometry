```python
import lda
import pandas as pd
import numpy as np
import lda.datasets
import re
```

```python
data = pd.read_csv("./datasets/DocumentTermMatrixcorsican_withoutSW_dataset.tsv", '\t')
```

```python
data = data.set_index('Unnamed: 0')
data
```

```python
data = data.T
```

```python
data_mat = data.to_numpy()
```

```python
data_mat.shape
```

```python
data_mat.sum()
```

```python
model = lda.LDA(n_topics=20, n_iter=1500, random_state=1)
```

```python
model.fit(data_mat)
```

```python
topic_word = model.topic_word_
```

```python
n_top_words = 8
```

```python
with open("./datasets/vocabulary_corsican_withoutSW.txt", 'r') as f:
    content = f.readlines()
```

```python
vocab =  []
for i in content:
    vocab.append(re.sub('\n', '', i))
```

```python
vocab = tuple(sorted(vocab))
```

```python
dataset = pd.read_csv('./datasets/full_corsican_withoutSW_dataset.tsv', '\t')
```

```python
texts = tuple(dataset['Texts'].tolist())
```

```python
for i, topic_dist in enumerate(topic_word):
    topic_words = np.array(vocab)[np.argsort(topic_dist)][:-(n_top_words+1):-1]
    print('Topic {}: {}'.format(i, ' '.join(topic_words)))
```

```python
doc_topic = model.doc_topic_
```

```python
for i in range(10):
    print("{} (top topic: {})".format(texts[i], doc_topic[i].argmax()))
```

### Conclusion

Pas très convaincant, les résultats ne sont pas particulièrement logiques. Il fallait s'y attendre car cette méthode est efficace sur des corpus plus petits.
