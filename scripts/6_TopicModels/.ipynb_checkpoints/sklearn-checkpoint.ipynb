```python
review_file = './datasets/full_corsican_withoutSW_dataset.tsv'
```

```python
import pandas as pd
```

```python
df = pd.read_csv(review_file, '\t')
```

```python
df_type = df[df['Type'].str.contains("poe")]
df_aut = df[df['Auteurs'].str.contains("Petru Rocca")]
x_train = df_type['Texts'].tolist()
```

```python
co_stops = ["li","le","e","u","a","la","duve","pa","ste", "pa","parchi","sempri","anche","ancu","unn","ellu","elli","ella","elle","quellu","quelle","quelli","quella","so","aghju","aghiu","he","simu","avemu","site","hannu","hanu","pe","par","seraghiu","sera","sara","oramai","tuttu","chi","quessu","quessa","quesse","quessi","bellu","bonu","bona","bon","boni","bone","boni","be","benche","mo", "lu", "cu","incu","nostru","vostru",'cusi','cun','st','bi', 'micca', 'altru','to','avia', 'stu', 'quandu', 'dopu','ca', 'ava', 'sottu', 'pocu','tutt', 'ind', 'inde', 'unu', 'tantu', 'dui','oh', 'fattu', 'vo', 'eiu', 'gran', 'bella', 'eccu', 'cum','nun', 'caru','cara','cari']
```

```python
import ssl
ssl._create_default_https_context = ssl._create_unverified_context

from sklearn.feature_extraction.text import CountVectorizer



tf_vectorizer = CountVectorizer(stop_words=co_stops, max_df=0.5, min_df=5,max_features = 7000, ngram_range=(1,2))
tf = tf_vectorizer.fit_transform(x_train)
```

```python
tf[0]
```

```python
from sklearn.decomposition import LatentDirichletAllocation

n_components = 8

lda = LatentDirichletAllocation(n_components=n_components, max_iter=20,
                                learning_method = 'online',
                                n_jobs=-1,verbose=1)
lda.fit(tf)
```

```python
def print_top_words(model, feature_names, n_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print()
        message = f'Topic {topic_idx}: '
        message += ', '.join([feature_names[i]
                             for i in topic.argsort()[:-n_top_words - 1:-1]])
        print(message)
    print()
```

```python
n_top_words = 20
tf_feature_names = tf_vectorizer.get_feature_names()
print_top_words(lda, tf_feature_names, n_top_words)
```

```python
for al, topic in enumerate(lda.components_):
    print(f"Topic {al} -> {topic}")
```

```python
# Autre manière de montrer les Topics

for i, comp in enumerate(lda.components_):
    vocab_comp = zip(tf_feature_names, comp)
    sorted_words = sorted(vocab_comp, key= lambda x:x[1], reverse=True)[:20]
    compp = sorted(comp, reverse=True)
    print("Topic "+str(i)+": ")
    n = 0
    for t in sorted_words:
        n += 1
        print(f"{t[0]}, {compp[n]}" ,end=" | ")
    print("\n")
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
def draw_word_cloud(index):
    imp_words_topic=""
    comp=lda.components_[index]
    vocab_comp = zip(tf_feature_names, comp)
    sorted_words = sorted(vocab_comp, key= lambda x:x[1], reverse=True)[:50]
    for word in sorted_words:
        imp_words_topic=imp_words_topic+" "+word[0]
        

    wordcloud = WordCloud(width=600, height=600, background_color='white').generate(imp_words_topic)
    #y = plt.figure( figsize=(5,5))
    #y = plt.imshow(wordcloud)
    #y = plt.axis("off")
    #y = plt.tight_layout()
    return wordcloud
```

```python
fig = plt.figure(figsize=(20,10), dpi=150)
for i in range(8):
    ax = fig.add_subplot(2, 4 ,i+1)
    wordcloud = draw_word_cloud(i)
    
    titre = f"Topic {i}"
    ax.imshow(wordcloud)
    ax.axis('off')
    ax.set_title(titre, fontsize=18)


```
