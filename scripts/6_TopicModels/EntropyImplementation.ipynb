### Entropy

#### Preprocessing

```python
import nltk
import pandas as pd
import numpy as np
import math
import itertools
import operator as op
import re

from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(r'\w+')
```

```python
corpus = [x.lower() for x in docs]
```

```python
tokens = [tokenizer.tokenize(x) for x in corpus]
```

```python
frequence = dict(nltk.FreqDist(list(itertools.chain.from_iterable(tokens))))
```

```python
#frequence = {key:val for key, val in frequence.items() if val > 1}
```

```python
vocabulaire = list(frequence.keys())
```

```python
def probability(i, j):
    tf = op.countOf(j, i)
    gf = frequence[i]
    return tf/gf, gf
```

```python
number_of_docs = len(corpus)
```

#### Tests

#### Entropy in the corpus

```python
dic = {}
for mot in vocabulaire:
    probs = []
    
    for doc in tokens:
        prob, freq = probability(mot, doc)
        probs.append(prob)
        
    count = [probs.index(x) for x in probs if x > 0]
    # Classification Token-Entropy
    
    entropie = -sum([probs[i] * np.log2(probs[i]) for i in count])/np.log2(number_of_docs)
    
    if entropie == -0.0:
        entropie = 0
        
    print(f"The entropy of the word « {mot} » is {entropie}")
    
    weight = freq * entropie
    
    print(f"The weight of the word « {mot} » is {weight}")
    
    dic[mot] = [entropie, weight]
```

```python
classification = pd.DataFrame(data=dic, index=["Entropy", "Weight score"]).T
```

```python
classification.to_csv("entropy_score.tsv", "\t")
```
