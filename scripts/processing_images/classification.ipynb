```python
import os
```

```python
parent = os.getcwd()
print('Directory : ' + parent)
```

```python
ark_folders = os.listdir()  #liste qui contient le nom de tous les dossiers -> correspond aux identifiants ARK
ark_folders.remove('pages_1') #suppression des dossiers 'réceptacles' de la liste
ark_folders.remove('pages_2')
ark_folders.remove('pages_3')
ark_folders.remove('pages_4')
ark_folders.remove('.ipynb_checkpoints')

print(ark_folders)
```

```python
for name in ark_folders: #on supprime de la liste tous les éléments qui ne sont pas des dossiers
    if os.path.isdir(name): #afin de retenir seulement les codes ARK
        print(name)
    else:
        ark_folders.remove(name)
```

```python
os.chdir(parent + '/pages_1/out') #Enfin, on se déplace dans les dossiers outpu de scan_tailor pour déplacer
filenames = os.listdir() #les fichiers .tif dans le dossier correspondant en se basant sur le code ARK
for name in ark_folders:
    for file in filenames:
        if file.endswith(name + '_p1.tif'):
            os.rename(file, parent + '/' + name + '/' + file)
```
